Task 1 : Sentiment Analysis


```python
from transformers import pipeline

# sentiment analysis using english

nlp = pipeline("sentiment-analysis")

result = nlp("I hate you")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")

result = nlp("I love you")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

```python
# sentiment analysis using arabic and french

nlp = pipeline("sentiment-analysis")

result = nlp("je suis triste")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")

result = nlp("انا حزين")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

```python
nlp = pipeline("sentiment-analysis", model="bert-base-multilingual-cased")

result = nlp("je suis très triste")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")

result = nlp(" انا سعيد")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

```python

nlp = pipeline("sentiment-analysis", model="bert-base-multilingual-uncased", tokenizer="bert-base-uncased")

result = nlp("je suis très triste")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")

result = nlp(" انا سعيد")[0]
print(f"label: {result['label']}, with score: {round(result['score'], 4)}")
```

Task 3 : Name entity recognition (NER)

```python
# using english

nlp = pipeline("ner")

sequence = "Hugging Face Inc. is a company based in New York City. Its headquarters are in DUMBO, therefore very"
          
print(nlp(sequence))
```

```python
# using arabic

sequence = "قمين عنوان شفرة عنوان بالنظر إلى عدد صحيح الصفيف وهدف القيمة المستهدف، يرجى الاطلاع على الأعداد الصحيحة وإرجاعها للقيمة المستهدفة في هذه الصفيف والعودة إلى مجموعة صفيفها"
          
print(nlp(sequence))
```

Task 4 : Question answering

```python
# using english

nlp = pipeline("question-answering")

context = "Extractive Question Answering is the task of extracting an answer from a text given a question. An example of a question answering dataset is the SQuAD dataset, which is entirely based on that task."


print(nlp(question="What is extractive question answering?", context=context))
print(nlp(question="What is a good example of a question answering dataset?", context=context))
```

```python
# using arabic

from transformers import pipeline
nlp = pipeline("question-answering")

context="مرحبا انا اسمي امال"

print(nlp(question="ما هو اسمك ؟", context=context))
```

Task 6 : Summarization

```python
# using arabic

summarizer = pipeline("summarization")

ARTICLE = "ويكيبيديا مشروع تعاوني متعدد اللغات يضم ويكيات بأكثر من 300 لغة للعمل في مشاريع موسوعات حرة ودقيقة ومتكاملة ومتنوعة ومحايدة، يستطيع الجميع المساهمة في تحريرها. نشأت ويكيبيديا في عام 2001، حيث نمت وتطورت بسرعة لتصبح واحدة من أكبر المواقع على الإنترنت. بدأت النسخة العربية في يوليو/تموز 2003."
print(summarizer(ARTICLE, max_length=130, min_length=30, do_sample=False))
```

```python
# using french

summarizer = pipeline("summarization")
ARTICLE="Wikipédia est un projet d’encyclopédie collective en ligne, universelle, multilingue et fonctionnant sur le principe du wiki. Ce projet vise à offrir un contenu librement réutilisable, objectif et vérifiable, que chacun peut modifier et améliorer."
print(summarizer(ARTICLE, max_length=84, min_length=30, do_sample=False))
```
